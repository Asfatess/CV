Задание 17. Выполните стыковку тензоров a и b по вертикали.

Вход:

> a = torch.arange(2, 10, 2).reshape(2, -1)

> b = torch.repeat(1, 10).reshape(2, -1)

Желаемый результат: новый тензор размером $4 \times 5$.


```python
# В задании ошибка, нет функции torch.repeat.
# Думаю, можно заполнить тензор любыми элементами (например единицами) и просто состыковать любые 2 тензора

a = torch.arange(2, 10, 2).reshape(2, -1)
b = torch.ones(10).reshape(2, -1)
c = torch.zeros(2, 5)
c[:, :2] = a
d = b.view(2, 5)

result = torch.cat([c, d])
print(result)
```

Задание 18. Выполните стыковку тензоров a и b по горизонтали.

Вход:

a = torch.arange(2, 10, 2).reshape(2, -1)

b = torch.repeat(1, 10).reshape(2, -1)

Желаемый результат: новый тензор размером $2 \times 10$.

```python
# Здесь встречается та же опечатка с repeat
# repeat заменена на ones
a = torch.arange(2, 10, 2).reshape(2, -1)
b = torch.ones(10).reshape(2, -1)
tensor8 = torch.cat([a, b], dim=1)
print(tensor8)
```

Задание 19. Получите все элементы от 5 до 10 в тензоре a.

Вход:

> a = torch.tensor([2, 6, 1, 9, 10, 3, 27])

Желаемый результат:

> tensor([6, 9, 10])

```python
a = torch.tensor([2, 6, 1, 9, 10, 3, 27])
tensor9 = a[(a >= 5) & (a <= 10)]
print(tensor9)
```
